```python
import thulac
import jieba
import pandas as pd
import xlwings as xw

posts = list(pd.read_csv('dataclean2.csv')['Post'].values)
labels = list(pd.read_csv('dataclean2.csv')['Label'].values)

thu1 = thulac.thulac(model_path='./Thulac_models/', user_dict='./user_dict.txt', rm_space=True, seg_only=True, filt=True)
jieba.load_userdict('./user_dict.txt')
```

```python
segement_posts_thulac = []
# segement_posts_jieba = []
for post in posts:
    if type(post) == str:
        t1 = thu1.cut(post, text=True) 
#         t2 = ' '.join(jieba.cut(post))
        segement_posts_thulac.append(t1)
#         segement_posts_jieba.append(t2)


df = pd.DataFrame({'segement': segement_posts_thulac})
df.to_excel("segementation.xlsx", encoding='utf-8')  
# wb = xw.Book()
# sht = wb.sheets.add(name='thulac')
# sht.range('A1').value = segement_posts_thulac
# sht = wb.sheets.add(name='jieba')
# sht.range('A1').value = segement_posts_jieba
# wb.save('segement_posts.xlsx')
```

```python
# get all the emoji text codes
import re

with open('./emoji_unicodes.txt', 'r') as f:
    lines = f.readlines()

emoji_codes = []
for line in lines:
    pattern = re.compile(":(.*):")
    result = re.findall(pattern, line)
    if result:
        emoji_code = ':' + result[0][:-1] + '\n'
        emoji_codes.append(emoji_code)

with open('./user_dict.txt', 'w') as f:
    f.writelines(emoji_codes)
```

```python
# transfer LIWC_ch dict to user_dict
import re

with open('./Simplified_Chinese_LIWC2015_Dictionary.dic', 'r') as f:
    lines = f.readlines()
    
word_lines = lines[83:]

words = []
for word_line in word_lines:
    pattern = re.compile('(.*?)[*\t]')
    result = re.findall(pattern, word_line)
    word = result[0] + '\n'
    words.append(word)

with open('./user_dict.txt', 'a') as f:
    f.writelines(words)
    
```
